```python
import scipy.linalg as slin
```

```python
a = np.array([[ 3, -5, 8],

[-1, 2, 3],

[-5, -6, 2]])
```

# determination

```python
slin.det(a)
```

# trace

```python
np.trace(a)
```

# inverse

```python
slin.inv(a)
```

# transpose

```python
np.transpose(a)
```

# image pattern

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as slin
```

```python
image = np.random.randint(0,255, size=[255,255])


plt.plot(image)
plt.show()

```

```python
image = np.random.randint(0,255, size=[50,50])
plt.matshow(image, cmap="gray")
plt.show()
```

```python
image = np.array([[ 3, -5, 8],

[-1, 2, 3],

[-5, -6, 2]])

image = slin.inv(image)

plt.matshow(image)
plt.show
```
