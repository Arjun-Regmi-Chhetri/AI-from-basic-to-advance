```python
import scipy.linalg as slin
```

```python
a = np.array([[ 3, -5, 8],

[-1, 2, 3],

[-5, -6, 2]])
```

# determination

```python
slin.det(a)
```

# trace

```python
np.trace(a)
```

# inverse

```python
slin.inv(a)
```

# transpose

```python
np.transpose(a)
```
