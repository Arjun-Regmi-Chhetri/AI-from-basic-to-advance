# array using numpy

```python
import numpy as np
```

```python
x = np.array([0,6])
y = np.array([0,25])

plt.plot(x,y)
plt.show()
```

# adding title

```python
plt.plot(x,y)
plt.title("Data")
plt.show()
```

```python
plt.plot(x,y, marker='o')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.title("Data")
plt.show()
```

```python
plt.plot(x,y, 'o:r')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.title("Data")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([3, 8, 1, 10])


plt.plot(x, 'o:b', ms= 20, mec = 'r')
plt.show()
```

```python
x = np.array([3, 8, 1, 10])


plt.plot(x, marker="o", ms= 20, mec = 'b', mfc="r")
plt.show()
```

```python
x = np.array([3, 8, 1, 10])


plt.plot(x, marker="o", ms= 20, mec = 'r', mfc="r")
plt.show()
```
