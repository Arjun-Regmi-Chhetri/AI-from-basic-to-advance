```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression as LR
from sklearn.preprocessing import PolynomialFeatures as PF
from sklearn.linear_model import LogisticRegression as LGR
```

```python
data = pd.read_csv("../position_salary.csv")
data
```

```python
data = data[['Level', 'Salary']]
data = data.dropna()
data
```

```python
x = data.iloc[:, 0:1].values
y = data.iloc[:, 1:].values
```

```python
x
```

```python
y
```

```python
# Fitting Linear Regression to the dataset
lin = LR()
lin.fit(x,y)
```

```python
#Visualizing Linear Regression Results
plt.scatter(x,y, color = "cyan")
plt.plot(x, lin.predict(x), color="magenta")
plt.title("Linear Regression")
plt.xlabel("Position")
plt.ylabel("Salary")
plt.show()
```

```python
# Fitting Polynomial Regression to the dataset
pol_reg = PF( degree= 7)

# poly_reg.fit(X, y)
x_poly = pol_reg.fit_transform(x)

poly_reg = LR()

poly_reg.fit(x_poly, y)
```

```python
# Visualizing the Polymonial Regression results
plt.scatter(x,y, color = "cyan")
plt.plot(x, poly_reg.predict(x_poly), color="magenta")
plt.title("Truth or Bluff (Linear Regression)")
plt.xlabel("Position")
plt.ylabel("Salary")
plt.show()
```

```python
# Predicting a new result with Linear Regression
print(lin.predict([[6]]))


# Predicting a new result with Polymonial Regression
print(poly_reg.predict(pol_reg.fit_transform([[8]])))
```

```python
# Fitting Linear Regression to the dataset
lr = LGR()
lr.fit(x,y)
```

```python
# Visualizing the Polymonial Regression results
plt.scatter(x,y, color = "cyan")
plt.plot(x, lr.predict(x), color="magenta")
plt.title(" Logistic Regression ")
plt.xlabel("Position")
plt.ylabel("Salary")
plt.show()
```
