```python
# descending order by index

df = pd.read_csv("book.csv", nrows=8)
df.sort_index(axis=0, ascending=False)
```

# Note -----------------------------------

  * axis = 0 means sorting through rows
  * axis = 1 means sorting through columns

# ----------------------------------------

```python
df.sort_index(axis=1, ascending=False)
```

```python
df.sort_index(axis=1)
```

# modification  ------------------------------------------------------------------------------------------

```python
df = pd.read_csv("book.csv")
df
```

```python
# Changing 11 row Title

df["Title"][11] = "Hello Python"

df.loc[11]
```

# Note

* though above change the data is not considered the ideal way

```python
df.loc[11, "Title"] = "Hello Python"
```

```python
df[df["Title"] == "Hello Python"]
```

```python
# change publication year

df.loc[11, "Publication Year"] = 2024
```

```python
df[df["Title"] == "Hello Python"]
```

```python
df.iloc[[11]]
```
