# Read / Write Multiple dataframe

````python
list_1 = {'Names': ['Andreas', 'George', 'Steve',

'Sarah', 'Joanna', 'Hanna'],

'Age':[21, 22, 20, 19, 18, 23]}



list_2 = {'Names': ['Pete', 'Jordan', 'Gustaf',

'Sophie', 'Sally', 'Simone'],

'Age':[22, 21, 19, 19, 29, 21]}


list_3 = {'Names': ['Ulrich', 'Donald', 'Jon',

'Jessica', 'Elisabeth', 'Diana'],

'Age':[21, 21, 20, 19, 19, 22]}
````

````python
df_lists = []
for item in [list_1, list_2, list_3]:
    df_list = pd.DataFrame(item)
    df_lists.append(df_list)


df = pd.concat(df_lists, keys =['Group1', 'Group2', 'Group3'], names=['Group', 'Row Num']).reset_index()

df.to_csv("multiframe.csv", index=False)
````
