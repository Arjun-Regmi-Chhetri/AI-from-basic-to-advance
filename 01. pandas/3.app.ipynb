# iloc()

* **It is used to access rows and columns**

```python
df = pd.read_csv("book.csv")
df.iloc[0:4, 0:]
```

```python
df.iloc[10:15, 0:3]
```

```python
df.iloc[:-1, 0:]   # it exclude last item
```

# loc vs iloc

* loc return specific row
* iloc return specific row and colum
